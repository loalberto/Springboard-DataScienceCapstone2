# 1b. Statistical Inference<a id ='1b.Statistical_Inference'></a>

Null Hypothesis: if it is possible to get an amount higher than $500 then we will always get something close to that amount.

Alternative Hypothesis: The value of total amount will always remain near the mean.

```python
# First is to check whether the data for 'total_amount' is normally distributed.
total_amount_normal = stats.normaltest(df['total_amount'])
total_amount_normal
# A p-value of 0.0 tells us that it is not normally distributed.
```

```python
# The histrogram provides further proof that the data is not normally distributed
plt.hist(df['total_amount'], bins=10)
plt.show()
```

The distribution of the data is not normalized. So a permuatation test can be used here to determine the p-value and confidence interval.

```python
vals = np.random.permutation(df['total_amount'])
temp_df = pd.DataFrame({'total_amount': df['total_amount'],
                       'Permutation_values': vals})
temp_df.head()
```
