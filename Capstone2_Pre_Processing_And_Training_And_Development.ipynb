```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('./datasets/Crucial_Elements.csv', index_col=0)
df = df.drop(df.index[0], axis=0)
df.tail()
```

I created dummy variables in my data wrangling to help study what the data looks like and to check how they correlated to the target variable.

# Standardizing the magnitude

```python
scalar = StandardScaler()
scaled_df = scalar.fit_transform(df)
scaled_df = pd.DataFrame(scaled_df, columns=df.columns)
scaled_df.head()
```

```python
len(df.columns)
```

```python
scalar = StandardScaler()
indecies = df.index
X = df.drop('total_amount', axis=1)
y = df['total_amount']
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7)
X_train.head()

# Join X_train and y_train into one dataframe
training_data = X_train.join(y_train)
# Same with the testing data
testing_data = X_test.join(y_test)

# Fit on the training data 
scalar.fit(training_data)

# Transform on the training and testing data
training_data = scalar.transform(training_data)
testing_data = scalar.transform(testing_data)
training_data = pd.DataFrame(training_data, columns=df.columns, index=indecies[:len(X_train)])
testing_data = pd.DataFrame(testing_data, columns=df.columns, index=indecies[len(X_train):])
testing_data.tail()
```

```python
training_data.to_csv('./datasets/Train_Data.csv')
testing_data.to_csv('./datasets/Test_Data.csv')
```
